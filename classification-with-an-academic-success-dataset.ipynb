```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from keras.utils import to_categorical
```

```python
train=pd.read_csv("/kaggle/input/playground-series-s4e6/train.csv")
test=pd.read_csv("/kaggle/input/playground-series-s4e6/test.csv")
```

```python
x_train=train.drop(['Target','id'],axis="columns")
```

```python

y_train=train["Target"]
y_train.unique()
```

```python
encoder = LabelEncoder()
y_train=encoder.fit_transform(y_train)
y_train
```

```python
x_test=test.drop("id",axis="columns")
```

```python

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
df_train = pd.DataFrame(x_train_scaled, columns=x_train.columns)
df_test = pd.DataFrame(x_test_scaled, columns=x_test.columns)
```

```python
bestfeatures = SelectKBest(score_func=chi2, k=15)
fit = bestfeatures.fit(df_train,y_train)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(df_train.columns)
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Specs','Score']
```

```python
features=featureScores.nlargest(15,'Score') 
features
```

```python
Selected_x_train=df_train
for i in df_train.columns:
    if i in features["Specs"].values:
        print(i)
    else:
        Selected_x_train=Selected_x_train.drop(i,axis="columns")
```

```python
Selected_x_test=df_test
for i in df_test.columns:
    if i in features["Specs"].values:
        print(i)
    else:
        Selected_x_test=Selected_x_test.drop(i,axis="columns")
```

```python
Selected_x_test
```

```python
from catboost import CatBoostClassifier, Pool
catboostmodel = CatBoostClassifier(iterations=1000,
                           learning_rate=0.1,
                           depth=7,
                           eval_metric='Accuracy',
                           random_seed=42,
                           verbose=50)
catboostmodel.fit(Selected_x_train,y_train, plot=True)
```

```python
y_val_pred = catboostmodel.predict(Selected_x_test)
    
y_val_pred
```

```python
import numpy as np
original_classes = encoder.classes_
output=original_classes[y_val_pred]
output
output_df = pd.DataFrame(output, columns=['Target'])
```

```python
samp=pd.read_csv("/kaggle/input/playground-series-s4e6/sample_submission.csv")
datasets=pd.concat([samp["id"],output_df],axis=1)
datasets.columns=["id","Target"]
datasets.to_csv("sample_prediction_23.csv",index=False)
```

```python
import tensorflow as tp
from tensorflow import keras
from tensorflow.keras.layers import Dense, Dropout
model1=keras.Sequential([
    keras.layers.Dense(65,input_shape=(Selected_x_train.shape[1],),activation="relu"),
    Dropout(0.5),
    keras.layers.Dense(65,activation="relu"),
    Dropout(0.5),
    keras.layers.Dense(65,activation="relu"),
    Dropout(0.5),
    keras.layers.Dense(3,activation="softmax")

])
model1.compile(
    optimizer="sgd",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"])
```

```python
from tensorflow.keras.callbacks import EarlyStopping

early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

history=model1.fit(Selected_x_train,y_train,epochs=10,batch_size=32,callbacks=[early_stopping])
```

```python
history=model1.fit(Selected_x_train,y_train,epochs=10,batch_size=32,callbacks=[early_stopping])
```

```python
predicted=model1.predict(Selected_x_test)
```

```python
import numpy as np
max_indices = np.argmax(predicted, axis=1)
```

```python
original_classes = encoder.classes_
output=original_classes[max_indices]
output
```

```python
output_df = pd.DataFrame(output, columns=['Target'])
```

```python
samp=pd.read_csv("/kaggle/input/playground-series-s4e6/sample_submission.csv")
datasets=pd.concat([samp["id"],output_df],axis=1)
datasets.columns=["id","Target"]
datasets.to_csv("sample_prediction_16.csv",index=False)
```
