```python
import pandas as pd
import tensorflow 
from tensorflow import keras
from sklearn.preprocessing import MinMaxScaler
```

```python
df=pd.read_csv("/kaggle/input/multiple-linear-regression-dataset/multiple_linear_regression_dataset.csv")
```

```python
df
```

```python
x=df.drop("income",axis="columns")
y=df["income"]
```

```python
sx=MinMaxScaler()
sy=MinMaxScaler()
x=sx.fit_transform(x)
y=sy.fit_transform(y.values.reshape(df.shape[0],1))
```

```python
import numpy as np
number_of_features=x.shape[1]
w=np.ones(number_of_features)
w
```

```python
def gradient_descent(x,y,epochs,lr=0.01):
    number_of_features=x.shape[1]
    w=np.ones(number_of_features)
    b=0
    total_samples=x.shape[0]
    
    total_cost=[]
    epoch=[]
    
    for i in range(epochs):
        predict=np.dot(w,x.T)+b
        
        dw=-(2/total_samples)*(x.T.dot(y-predict))
        db=-(2/total_samples)*np.sum(y-predict)
        w=w-lr*dw
        b=b-lr*db
        print(dw)
        cost=np.mean(np.square(y-predict))
        if i%10==0:
            total_cost.append(cost)
            epoch.append(i)
    return w,b,cost,total_cost,epoch
    
```

```python
w,b,cost,total_cost,epoch=gradient_descent(x,y.reshape(y.shape[0]),500)
```

```python
def predict(age,experience,w,b):
    scaledx=sx.transform([[age,experience]])[0]
    scaled_price=w[0]*scaledx[0]+w[1]*scaledx[1]+b
    return sy.inverse_transform([[scaled_price]])[0][0]
print(w,b)
predict(39,8,w,b)
```

```python

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=5)
```

```python
model=keras.Sequential([
    keras.layers.Dense(3,input_shape=(X_train.shape[1],),activation="relu"),
    keras.layers.Dense(1)
    
])
model.compile(
optimizer="sgd",
loss="mean_squared_error",
metrics=["accuracy"])
```

```python
model.fit(X_train,y_train,epochs=500)
```

```python
ypred=model.predict(X_test)
```

```python
ypred=sy.inverse_transform(ypred)
```
