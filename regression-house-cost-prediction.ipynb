```python
import tensorflow as tf
import pandas as pd
from sklearn.preprocessing import LabelEncoder,MinMaxScaler
from tensorflow import keras 
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("/kaggle/input/regression-dataset-for-household-income-analysis/data.csv")
```

```python
df
```

```python
for columns in df:
    print(columns,len(df[columns].unique()))
```

```python
columns=["Education_Level","Occupation","Location","Marital_Status","Employment_Status","Homeownership_Status","Type_of_Housing","Gender","Primary_Mode_of_Transportation"]
```

```python
df=pd.get_dummies(df).astype(int)
```

```python

num_columns=["Age","Number_of_Dependents","Work_Experience","Household_Size"]
xs=MinMaxScaler()
df[num_columns]=xs.fit_transform(df[num_columns])
```

```python
df
```

```python
x=df.drop("Income",axis=1)
y=df["Income"]
```

```python
sy=MinMaxScaler()
y=sy.fit_transform(y.values.reshape(df.shape[0],1))
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
import tensorflow as tp
from tensorflow import keras
from tensorflow.keras.layers import Dense, Dropout
model=keras.Sequential([
    keras.layers.Dense(32,input_shape=(x_train.shape[1],),activation="relu"),
    keras.layers.Dense(15,activation="relu"),
    keras.layers.Dense(1)

])
model.compile(
    optimizer="sgd",
    loss="mean_squared_error",
    metrics=[keras.metrics.RootMeanSquaredError()])
```

```python
model.fit(x_train,y_train,epochs=250)
```

```python
y_pred=model.predict(x_test)
```
