```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd
```

```python
test=pd.read_csv("/content/test.csv")
```

```python
train=pd.read_csv("/content/train.csv")
```

```python
df_train =train.drop("price",axis="columns")
```

```python
df_test =test
```

```python
encoder = LabelEncoder()
```

```python
combined = pd.concat([df_train, df_test])
```

```python
columns=["brand","accident","clean_title","model","fuel_type","engine","transmission","ext_col","int_col"]
for column in columns:
    # Fit the encoder on the combined data
    encoder.fit(combined[column])

    # Transform train and test data separately
    df_train[column] = encoder.transform(df_train[column])
    df_test[column] = encoder.transform(df_test[column])

print("Encoded Train Data:")
print(df_train)
print("\nEncoded Test Data:")
print(df_test)
```

```python
col=["brand","model","model_year","milage","fuel_type","engine","transmission","ext_col","int_col"]
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_train[col]=scaler.fit_transform(df_train[col])
df_test[col]=scaler.fit_transform(df_test[col])
```

```python
df_train.drop("id",axis="columns")
df_test.drop("id",axis="columns")
```
